```python
import pandas as pd
import methods as met
import numpy as np
import matplotlib.pyplot as plt

file_path = 'gym_members_exercise_tracking.csv'
data = pd.read_csv(file_path)
# = 'Workout_TypeHIIT', 'Workout_TypeStrength', 'Workout_TypeYoga',  'Water_Intake..liters.', 'Workout_Frequency..days.week.'
data.drop(columns=['Max_BPM', 'Avg_BPM', 'Resting_BPM', 'Fat_Percentage',
                    'Experience_Level', 'Water_Intake (liters)', 'Workout_Frequency (days/week)'], inplace = True)
#print(data)

data_size = len(data['Age'])

#data partitioning
Male_data = data[data['Gender'] == 'Male']
Female_data = data[data['Gender'] == 'Female']


fig, ax = plt.subplots(2,3, figsize = (20, 10))
#response: weight

ax_flat = ax.flatten()
col_names = Male_data.columns
print(col_names)
k = 0
for i in ax_flat:
    i.scatter(Male_data[col_names[k]], Male_data['Weight (kg)'], color = "blue", label = "Male", alpha = 0.5)
    i.scatter(Female_data[col_names[k]], Female_data['Weight (kg)'], color = "red", label = "Female", alpha = 0.5)
    i.set_xlabel(col_names[k])
    i.set_ylabel("Weight (kg)")
    i.legend()
    i.set_title("Weight (kg) against "+ col_names[k])
    k += 1






#fig, ax = plt.subplots(2,3, figsize = (10,10))
#response: weight

#for i in ax:
#    i.scatter(Male_data['Age'], Male_data['Weight (kg)'], color = "blue", label = "Male", alpha = 0.5)




```
