```python
import pandas as pd
import methods as met
import numpy as np
import matplotlib.pyplot as plt

file_path = '500_height_weight.csv'
data = pd.read_csv(file_path)

data_set_size = 500
Male_height = []
Male_weight = []
Female_height = []
Female_weight = []

number_of_Males = 0
number_of_Females = 0

for i in range(data_set_size):
    if data['Gender'][i] == 'Male':
        Male_height.append(data['Height'][i])
        Male_weight.append(data['Weight'][i])
        number_of_Males += 1
    else:
        Female_height.append(data['Height'][i])
        Female_weight.append(data['Weight'][i])

number_of_Females = data_set_size - number_of_Males 

Male_data = np.zeros((number_of_Males, 2))
Female_data = np.zeros((number_of_Females, 2))
Male_data[:, 0] = Male_height
Male_data[:, 1] = Male_weight

Female_data[:, 0] = Female_height
Female_data[:, 1] = Female_weight

#Training = Males = de
#Test = Females = nu


plt.hist(Male_height, alpha = 0.5, bins = 30, label = "Male height distribution")
plt.hist(Female_height, alpha = 0.5, bins = 30, label = "Female height distribution")
avg_male_height = np.average(Male_height)
avg_female_height = np.average(Female_height)
plt.vlines(avg_male_height, ymin = 0, ymax = 18, label = "Average Male height", color = "red", linestyle = "--")
plt.vlines(avg_female_height, ymin = 0, ymax = 18, label = "Average Female height", color = "green")
estimation = met.KMM_simple(Male_height, Female_height)
plt.legend()
#plt.scatter(Male_height, estimation)
#met.KMM_simple(Male_height)

```
